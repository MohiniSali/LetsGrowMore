```python
import cv2
```

```python
#read the image
img=cv2.imread(r"C:\Users\Lenovo\Downloads\images.jpg")
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
#conver image to grayscale image
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis("off")
plt.imshow(gray_img)
plt.show()
```

```python
#invert the image
inv_gray_img=cv2.bitwise_not(gray_img)
plt.axis("off")
plt.imshow(inv_gray_img)
plt.show()
```

```python
#blur the image
blur_img=cv2.GaussianBlur(inv_gray_img,(5,5),0)
plt.axis("off")
plt.imshow(blur_img)
plt.show()
```

```python
#invert the image
inv_blur_img=cv2.bitwise_not(blur_img)
plt.axis("off")
plt.imshow(inv_blur_img)
plt.show()
```

```python
#creat pencil sketch image
pencil_img=cv2.divide(gray_img,inv_blur_img,scale=256.0)
plt.axis("off")
plt.imshow(pencil_img)
plt.show()
```

```python
#save the sketch
cv2.imwrite("./output_image.png",pencil_img)
```

```python
#display original image vs sketch
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title("image",size=18)
plt.imshow(img)
plt.axis("off")
plt.subplot(1,2,2)
plt.title("sketch image",size=18)
rgb_sketch=cv2.cvtColor(pencil_img,cv2.COLOR_BGR2RGB)
plt.imshow(pencil_img)
plt.axis("off")
plt.show()
```
